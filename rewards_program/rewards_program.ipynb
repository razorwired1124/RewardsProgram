```python
import pandas as pd
data = pd.read_csv('customer_info.csv')

#print(data)

def reward_points (row):
    if row['Purchase Amount'] > 50 and row['Purchase Amount'] <= 100 :
        return row['Purchase Amount'] - 50
    elif row['Purchase Amount'] > 100 :
        return (row['Purchase Amount'] - 100) * 2 + 50
    else:
        return 0
data['Reward Points'] = data.apply (lambda row: reward_points(row), axis=1)
print(data)
```
